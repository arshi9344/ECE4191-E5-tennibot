```python
import multiprocessing as mp
import time
import os
import psutil

from cpu_bound_task import cpu_bound_task

def main():
    # Get the number of CPU cores
    num_cores = mp.cpu_count()
    print(f"Number of CPU cores: {num_cores}")

    # Create a list to hold our processes
    processes = []

    # Spawn processes equal to the number of cores
    for i in range(num_cores):
        p = mp.Process(target=cpu_bound_task, args=(f"Process-{i}",))
        processes.append(p)
        p.start()

    # Wait for all processes to complete
    for p in processes:
        p.join()

if __name__ == "__main__":
    main()
```
