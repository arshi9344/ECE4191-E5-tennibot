# Code for line detection:  

```python
import time
import board
import busio
from IPython import display
import adafruit_tcs34725
from adafruit_extended_bus import ExtendedI2C as I2C

# Use I2C bus 3 with custom SDA and SCL pins
i2c = I2C(3)  # This specifies I2C bus 3

# Initialize the TCS34725 sensor
sensor = adafruit_tcs34725.TCS34725(i2c)

sensor.integration_time = 200
sensor.gain = 60

# Variable to change upon detecting a color shift
color_detected = 0

# Set initial threshold values for color change detection
threshold = 1000  # Adjust this value based on sensitivity requirements

# Store the initial color values as a reference
initial_r, initial_g, initial_b, _ = sensor.color_raw

print("Monitoring color changes...")

while True:
    # Read the current color values
    r, g, b, c = sensor.color_raw

    # Calculate the difference in color values
    r_diff = abs(r - initial_r)
    g_diff = abs(g - initial_g)
    b_diff = abs(b - initial_b)

    # Detect a color shift based on the threshold
    if r_diff > threshold or g_diff > threshold or b_diff > threshold:
        color_detected = 1
        print("Color shift detected! Variable set to 1.")
    else:
        color_detected = 0

    # Print current status (optional)
    print(f"Color detected: R={r}, G={g}, B={b}, Variable={color_detected}")

    # Optional: Update initial color if you want to detect a shift back
    # initial_r, initial_g, initial_b = r, g, b

    # Small delay to avoid excessive CPU usage
    time.sleep(0.1)
```
